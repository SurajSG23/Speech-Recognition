###  **Importing libraries**

These are “tools” that add extra features to your Python program.

```python
import numpy as np
import matplotlib.pyplot as plt
```

* **numpy (np)** → used for math and working with numbers, especially arrays (lists of numbers).
* **matplotlib.pyplot (plt)** → used to make graphs and plots.

---

```python
import librosa
import librosa.display
```

* **librosa** → used for **audio and music processing** (like loading sound files, finding beats, frequencies, etc.).
* **librosa.display** → helps to **visualize** sound (like spectrograms — pictures of sound).

---

```python
import soundfile as sf
```

* **soundfile** → used to **read and write audio files** (like `.wav`, `.flac`, etc.).

---

```python
import speech_recognition as sr
```

* **speech_recognition** → used to **convert speech to text** using a microphone or audio file.

---

```python
from jiwer import wer, cer
```

* **jiwer** → a library that helps measure how accurate a speech recognition system is.

  * **wer** = *Word Error Rate* → compares how many words were wrong.
  * **cer** = *Character Error Rate* → compares how many letters/characters were wrong.

---

```python
from IPython.display import Audio
```

* Lets you **play audio directly inside Jupyter Notebook**.

---

```python
import whisper
```

* **whisper** → OpenAI’s **speech recognition model** (very powerful, converts speech to text).

---

```python
import csv
import os
import tempfile
import wave
```

* **csv** → for reading/writing spreadsheet-style files (rows and columns).
* **os** → for working with your computer’s file system (folders, paths).
* **tempfile** → for making temporary files while running a program.
* **wave** → for working with `.wav` sound files directly.

---

```python
from gtts import gTTS
```

* **gTTS (Google Text-to-Speech)** → converts **text into spoken voice/audio** using Google’s speech engine.
